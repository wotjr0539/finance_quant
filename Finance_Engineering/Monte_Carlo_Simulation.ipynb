### 몬테카를로 시뮬레이션
    주가 = 30달러, 기대수익률 9%, 변동성 20%, 한달간격으로 5년 시뮬레이션하기, 랜덤변수는 노말분포, 주가를 그려보기. 랜덤샘플이 바뀔때마다 주가변화를 관찰해보기. 

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
S0 = 30
n = 52*5
mu = 0.09
sigma = 0.20
e = (mu/52 + sigma*np.random.randn(n, 100 ) / math.sqrt(52) + 1)
S = S0 * e.cumprod(axis=0)
# vstack - 행으로 붙이기
S = np.vstack([np.ones((1, 100))*S0, S])
```

```python
plt.plot(S)
```
